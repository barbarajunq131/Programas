1. Quest√£o 1 (4 pontos): Para a vari√°vel aleat√≥ria ‚ÄúTem Internet‚Äù:

```python
# Quest√£o 1.a - Probabilidade de "Tem Internet = Sim"

p_internet = (df["TemInternet"] == "Sim").mean()
print("\033[92mQuest√£o 1.a\033[0m")
print("Probabilidade de ter Internet (p):", round(p_internet, 4))

# Quest√£o 1.b - Probabilidades Binomiais
# X ~ Binomial(n=20, p=p_internet)

n = 20
# P(X < 10) = P(X ‚â§ 9)
p_x_menor10 = binom.cdf(9, n, p_internet)

# P(12 < X ‚â§ 15) = P(X ‚â§ 15) - P(X ‚â§ 12)
p_x_12a15 = binom.cdf(15, n, p_internet) - binom.cdf(12, n, p_internet)

print("\033[92mQuest√£o 1.b\033[0m")
print("P(X < 10) =", round(p_x_menor10, 4))
print("P(12 < X ‚â§ 15) =", round(p_x_12a15, 4))

# Quest√£o 1.c - Esperan√ßa e Vari√¢ncia

E_X = n * p_internet
Var_X = n * p_internet * (1 - p_internet)
print("\033[92mQuest√£o 1.c\033[0m")
print("E[X] =", round(E_X, 4))
print("Var[X] =", round(Var_X, 4))
```

2. Quest√£o 2 (3 pontos): Sabendo que ùëç~ùëÅ(ùúá, ùúé 2), com ùúá = 0 e ùúé 2 = 1, calcule as probabilidades
abaixo:


```python
# Quest√£o 2.a - P(0 ‚â§ Z ‚â§ 1,73)

p_0_173 = norm.cdf(1.73) - norm.cdf(0)
print("\033[92mQuest√£o 2.a\033[0m")
print("P(0 ‚â§ Z ‚â§ 1.73) =", round(p_0_173, 4))

# Quest√£o 2.b - P(Z ‚â§ -1.73) e P(Z ‚â• 1.73)

p_menor = norm.cdf(-1.73)
p_maior = 1 - norm.cdf(1.73)
print("\033[92mQuest√£o 2.b\033[0m")
print("P(Z ‚â§ -1.73) =", round(p_menor, 4))
print("P(Z ‚â• 1.73) =", round(p_maior, 4))

# Quest√£o 2.c - P(0.47 ‚â§ Z ‚â§ 1.73)

p_047_173 = norm.cdf(1.73) - norm.cdf(0.47)
print("\033[92mQuest√£o 2.c\033[0m")
print("P(0.47 ‚â§ Z ‚â§ 1.73) =", round(p_047_173, 4))

```

3. Quest√£o 3 (3 pontos): Suponha que X seja igual a vari√°vel ‚ÄúRenda Antes‚Äù e tenha distribui√ß√£o
ùëÅ(ùúá, ùúé 2).

```python
# Quest√£o 3.a - Estimativas de m√©dia e desvio-padr√£o

mu_hat = df["RendaAntes"].mean()
sigma_hat = df["RendaAntes"].std()
print("\033[92mQuest√£o 3.a\033[0m")
print("M√©dia estimada (muÃÇ) =", round(mu_hat, 2))
print("Desvio padr√£o estimado (sigmaÃÇ) =", round(sigma_hat, 2))

# Quest√£o 3.b - P(X ‚â§ 1000)

p_x_menor1000 = norm.cdf(1000, loc=mu_hat, scale=sigma_hat)
print("\033[92mQuest√£o 3.b\033[0m")
print("P(X ‚â§ 1000) =", round(p_x_menor1000, 4))

# Quest√£o 3.c - P(X ‚â• 3000)

p_x_maior3000 = 1 - norm.cdf(3000, loc=mu_hat, scale=sigma_hat)
print("\033[92mQuest√£o 3.c\033[0m")
print("P(X ‚â• 3000) =", round(p_x_maior3000, 4))
```
