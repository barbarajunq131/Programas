Respostas aos exercicios da lista 02

1. Questão 1 (4 pontos): Para a variável aleatória “Tem Internet”:

```python
# Questão 1.a - Probabilidade de "Tem Internet = Sim"

p_internet = (df["TemInternet"] == "Sim").mean()
print("\033[92mQuestão 1.a\033[0m")
print("Probabilidade de ter Internet (p):", round(p_internet, 4))

# Questão 1.b - Probabilidades Binomiais
# X ~ Binomial(n=20, p=p_internet)

n = 20
# P(X < 10) = P(X ≤ 9)
p_x_menor10 = binom.cdf(9, n, p_internet)

# P(12 < X ≤ 15) = P(X ≤ 15) - P(X ≤ 12)
p_x_12a15 = binom.cdf(15, n, p_internet) - binom.cdf(12, n, p_internet)

print("\033[92mQuestão 1.b\033[0m")
print("P(X < 10) =", round(p_x_menor10, 4))
print("P(12 < X ≤ 15) =", round(p_x_12a15, 4))

# Questão 1.c - Esperança e Variância

E_X = n * p_internet
Var_X = n * p_internet * (1 - p_internet)
print("\033[92mQuestão 1.c\033[0m")
print("E[X] =", round(E_X, 4))
print("Var[X] =", round(Var_X, 4))
```

2. Questão 2 (3 pontos): Sabendo que 𝑍~𝑁(𝜇, 𝜎 2), com 𝜇 = 0 e 𝜎 2 = 1, calcule as probabilidades
abaixo:


```python
# Questão 2.a - P(0 ≤ Z ≤ 1,73)

p_0_173 = norm.cdf(1.73) - norm.cdf(0)
print("\033[92mQuestão 2.a\033[0m")
print("P(0 ≤ Z ≤ 1.73) =", round(p_0_173, 4))

# Questão 2.b - P(Z ≤ -1.73) e P(Z ≥ 1.73)

p_menor = norm.cdf(-1.73)
p_maior = 1 - norm.cdf(1.73)
print("\033[92mQuestão 2.b\033[0m")
print("P(Z ≤ -1.73) =", round(p_menor, 4))
print("P(Z ≥ 1.73) =", round(p_maior, 4))

# Questão 2.c - P(0.47 ≤ Z ≤ 1.73)

p_047_173 = norm.cdf(1.73) - norm.cdf(0.47)
print("\033[92mQuestão 2.c\033[0m")
print("P(0.47 ≤ Z ≤ 1.73) =", round(p_047_173, 4))

```

3. Questão 3 (3 pontos): Suponha que X seja igual a variável “Renda Antes” e tenha distribuição
𝑁(𝜇, 𝜎 2).

```python
# Questão 3.a - Estimativas de média e desvio-padrão

mu_hat = df["RendaAntes"].mean()
sigma_hat = df["RendaAntes"].std()
print("\033[92mQuestão 3.a\033[0m")
print("Média estimada (mû) =", round(mu_hat, 2))
print("Desvio padrão estimado (sigmâ) =", round(sigma_hat, 2))

# Questão 3.b - P(X ≤ 1000)

p_x_menor1000 = norm.cdf(1000, loc=mu_hat, scale=sigma_hat)
print("\033[92mQuestão 3.b\033[0m")
print("P(X ≤ 1000) =", round(p_x_menor1000, 4))

# Questão 3.c - P(X ≥ 3000)

p_x_maior3000 = 1 - norm.cdf(3000, loc=mu_hat, scale=sigma_hat)
print("\033[92mQuestão 3.c\033[0m")
print("P(X ≥ 3000) =", round(p_x_maior3000, 4))
```
